```python
import random

def createSuite(someList):
    for i in range(1,15):
        someList.append(i)
        
def createDeck(cardsList, suite1,suite2,suite3,suite4):
    cardsList.append(suite1)
    cardsList.append(suite2)
    cardsList.append(suite3)
    cardsList.append(suite4)
    
def drawSuite():
    return random.randint(0,3)
    
def drawNumber():
    return random.randint(1,14)

def resetVariables(handList,Fullhouse,handSuite,indexNum,cardNumb,Fullhouse1Num,fullHouse2Num):
    handList.clear()
    Fullhouse.clear()
    handSuite.clear()
    indexNum = 1
    cardNumb = 1
    Fullhouse1Num = 0
    Fullhouse2Num = 0
    
def printStats(FullHouseWin, Trials):
    print(f'Full House Stats: {fullHouseWin} / {trials} = {fullHouseWin/trials}')
    
def printStatsPair(pairs, Trials):
    print(f'Two Pair Stats: {pairs} / {Trials} = {pairs/Trials}')
    
def printStatsFlush(flush, Trials):
    print(f'Flush Stats: {flush} / {Trials} = {flush/Trials}')

```

```python
# #cards[x] = suite
# #cards[x][y] = number
suite = {0:"clubs", 1:"spades", 2:"diamonds",3:"hearts"}
cards = []
clubs = []
spades = []
diamonds = []
hearts = []

hand = {}
cardDrawn = ()
handSuite = []

# drawCards = 6

index = 1
cardNum = 1
fullHouse = []
fullHouse1 = 0
fullHouse2 = 0
fullHouseWin = 0
twoPairs = 0
flushWin = 0
bFlush = 1
flushCount = 0

createSuite(clubs)
createSuite(spades)
createSuite(diamonds)
createSuite(hearts)
createDeck(cards, clubs, spades, diamonds, hearts)

```

```python
## TRIALS AND PRINT CARDS HERE 

trials = 1000000

printCards = 0

## TRIALS AND PRINT CARDS HERE
```

```python
for i in range(0,trials):
    flushCount = 0
    bFlush = 1
    
    while len(hand) < 5:
        
        cardDrawn = (drawSuite(),drawNumber())
        
        if cardDrawn in hand.values():
            pass
        else:
            hand["card"+str(index)] = cardDrawn
            fullHouse.append(cardDrawn[1])
        
        index +=1
        
    # index = 0
    
    for j in range(0,len(fullHouse)):
        fullHouse1 = fullHouse[0]
        if fullHouse1 != fullHouse[j]:
            fullHouse2 = fullHouse[j]
            
    if fullHouse.count(fullHouse1) == 3:
        if fullHouse.count(fullHouse2) == 2:   
            if printCards == 1:
                print("Full House")
            fullHouseWin += 1
            
    elif fullHouse.count(fullHouse1)==2:
        if fullHouse.count(fullHouse2)==3:
            if printCards == 1:
                print("Full House")
            fullHouseWin += 1
        elif fullHouse.count(fullHouse2)==2:
            if printCards == 1:
                print("Two Pairs")
            twoPairs +=1
            
    if printCards == 1:           
        print(f'Hand {i}: ')

    for v in hand.values():
        for k,val in suite.items():     
            if v[0] == k:
                if printCards ==1:
                    print(f'Card {cardNum}: {v[1]} of {val}')
                handSuite.append(val)
                cardNum+=1
    cardNum = 1
    suiteVar = handSuite[0]

    for item in handSuite:
        if bFlush == 1:
            if item == suiteVar:
                flushCount +=1
        else: 
            bFlush = 0
    if flushCount == 5:
        flushWin +=1
            
        
    resetVariables(hand,fullHouse,handSuite,index,cardNum,fullHouse1,fullHouse2)
    
printStats(fullHouseWin, trials) 
printStatsPair(twoPairs,trials)
printStatsFlush(flushWin,trials)
fullHouseWin = 0
twoPairs = 0
flushWin = 0
# flushCount = 0
# bFlush = 1
```
